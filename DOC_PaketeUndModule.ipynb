# Funktionen von Paketen und Modulen einsehen

Mit Hilfe von Paketen und Modulen kann man den Funktionsumfang erhöhen bzw. den Programmieraufwand veringern. In Python werden Pakete und Module über den Befehl import am Anfang des Programms im Quellcode verfügbar gemacht.
Die Pakete und Module bieten dann Funktionen und Klassen die dann im Prgramm verwendet werden können.

Im Nachfolgenden Beispiel ist das package ipyturtle3 zunächst importiert und dann mit help() die im Paket eingebaute Hilfe aufgerufen:

```python
import ipyturtle3
help(ipyturtle3)
```
