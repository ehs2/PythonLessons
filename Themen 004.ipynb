## Bedingungen

Eins Bedingung ist ein Ausdruck der zwei (oder mehr) Werte mit einem Vergleichsoperator (wie >= oder <) verbindet und zu einem booleschen Wert auswertet. Eine Bedingung muss also immer zweifelsfrei True oder False ergeben
