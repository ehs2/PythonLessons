# Wiederholungen
Gegeben ist nachfolgende Grafik. Erläutern Sie offline mit Hilfe von Pseudo Code (vorwärts, rückwärts usw.) wie das Bild von Bob gezeichnet wurde.



```python
# Mit einem Import kann von anderen Programmiererinnnen erstellter Code
# in dem eigenen Programm genutzt werden

import ipyturtle3 as turtle
from ipyturtle3 import hold_canvas

#Eine Zeichenfläche/Lewinwand erzeugen
myCanvas=turtle.Canvas(width=500,height=500)
#Leinwand anzeigen
display(myCanvas)
myTS=turtle.TurtleScreen(myCanvas)
#Leinwandfarbe setzen
myTS.bgcolor("lightblue")
#myTS.bgcolor(122012)
myTS.clear()

#Bob wird als Schildkröte erzeugt und auf dem TurtleScreen angezeigt
bob=turtle.Turtle(myTS)

#Das Aussehen wird auf Schildkröte gesetzt
bob.shape("turtle")
#Die Geschwindigkeit der Turtle
myTS.delay(20)

#lassen Sie Bob das obige Sechseck zeichen
```

Das 6-Eck wird gezeichnet durch das 6-fache wiederholen der immergleichen Schritte: 60Grad drehen und 100 Pixel vorwärts.

Beim Programmieren sucht man immer nach geschickteren Lösungen, man möchte mit möglichst wenig Code auskommen.
Codewiederholungen sind auch deshalb zu verwmeiden, weil sich Fehler dann auch wiederholen.

Um den Code felxibler zu gestallten werden sogenannte Kontrollstrukturen verwenden.
Hier bietet sich eine Kontrollstruktur an die eine vorgege Anzahl von Schritten wiederholt: Die For-Schleife

```python
for i in range(6):
    bob.turnleft(60)
    bob.forward(80)
```
